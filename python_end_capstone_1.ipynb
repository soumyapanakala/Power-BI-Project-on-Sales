```python
#to read the data we need pandas library...
import pandas as pd

#now read the csv file and store in a variable...
data = pd.read_csv(r"C:\Users\yyy\Downloads\all_data.csv")
data
```

```python
#To remove the null values from the dataframe,first check whether there are null values..
data.isnull().sum()
```

```python
#Remove the null values 
data.dropna(inplace=True)
data.isnull().sum()    #now we can see that there are zero null values
```

```python
#Dataframe without null values is here..
data
```

```python
#Check whether there is any unwanted data or any duplicates...
data[data.duplicated()]
```

```python
#from the above we can see that there is unwanted data, so remove the unwanted data from the rows.. 
data.drop(data.index[(data["Product"] == "Product")],axis=0,inplace=True)
data
```

```python
#Now we can see that there is no such unwanted data..
data[data.duplicated()]
```

```python
#Data Preprocessing and EDA:
#Create Dependent column (Sales Column by multiplying Price eachColumn by Quantity ordered)
#To multiply the columns first check the datatypes, because if the datatypes are different error may come... 

data.info()          #As the datatyes are in object we can't multiply,so change the datatypes.
```

```python
data['Quantity Ordered']= pd.to_numeric(data['Quantity Ordered'])
```

```python
data['Price Each']=  pd.to_numeric(data['Price Each'])
```

```python
#Now we can easily multiply the columns and a new column 'Sales' is created

data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data                                                                 #we can see that sales column is created...
```

```python
#Create Multiple important Independent Column for better analysis like 
#                                                                   Month,Time,City,Product Category  etc...
```

```python
#Extract month column from 'Order Date'...
#First change the datatype of "Order Date" to datetime format
data["Order Date"] = pd.to_datetime(data["Order Date"],utc=True)
data.dtypes
```

```python
#now we can extract month from 'Order Date' by creating new column as 'Order Month'
data['Order Month']=data['Order Date'].dt.month_name(locale='English')
data
```

```python
#Now extract 'Order Time' from Order Date as new column name as 'Order Time'
data['Order Time']=data['Order Date'].dt.time
data
```

```python
#From 'Purchase Address' extract three independent columns ('Street','City','Zip Code') by splitting...
data[['Street','City','Zip Code']] = data['Purchase Address'].str.split(", ",expand=True)
data
```

```python
# As we dont want street and zip code columns we can remove those columns by using drop function...,

data.drop(columns = ['Street','Zip Code'], inplace = True)
data
```

```python
#From 'Product' column create column name as 'Product Category'.By using map function create this column...
data['Product'].unique()         #First check the unique values from the product column and then create product category column
```

```python
#Now for the unique items from product column, give categorical names and map it...,
Product_Category = {'USB-C Charging Cable':'Charging_cable', 'Bose SoundSport Headphones':'Head_phones',
       'Google Phone':'Mobile', 'Wired Headphones':'Head_phones', 'Macbook Pro Laptop':'Laptop',
       'Lightning Charging Cable':'Charging_cable', '27in 4K Gaming Monitor':'Monitor',
       'AA Batteries (4-pack)':'Batteries', 'Apple Airpods Headphones':'Head_phones',
       'AAA Batteries (4-pack)':'Batteries', 'iPhone':'Mobile', 'Flatscreen TV':'Television',
       '27in FHD Monitor':'Monitor', '20in Monitor':'Monitor', 'LG Dryer':'Dryer', 'ThinkPad Laptop':'Laptop',
       'Vareebadd Phone':'Mobile', 'LG Washing Machine':'Washing_machine', '34in Ultrawide Monitor':'Monitor'}
# After naming categories create 'Product Category column' by mapping...

data['Product Category'] = data['Product'].map(Product_Category)
data
```

```python
#FOR 'DATA VISUALIZATION' IMPORT THE TOOLS WE WANT...,

import matplotlib.pyplot as plt
import seaborn as sns

import ipywidgets as widgets
```

```python
#Relation between sales vs city using bargraph and line graph...,

c1 = widgets.Dropdown(
     options=['barplot', 'lineplot'],
     value='barplot',
     description='Choose graphtype:',
     disabled=False,
)

# Dropdown menu is to select either barplot or line plot as per reqirement----

def graphs(c2):
    if 'barplot'== c2:
        plt.figure(figsize=(10,4))
        sns.barplot(x= data['City'], y= data['Sales'], width = 0.5)
        plt.xlabel("CITY", fontsize = 18, color = 'blue')
        plt.ylabel("SALES", fontsize = 18, color = 'blue')
        plt.title("Sales vs City", fontsize = 20, color = 'red')
        plt.xticks(rotation = 45)
        plt.show()
        
    if 'lineplot'== c2:
        plt.figure(figsize=(10,4))
        sns.lineplot(x= data['City'], y= data['Sales'])
        plt.xlabel("CITY", fontsize = 18, color = 'blue')
        plt.ylabel("SALES", fontsize = 18, color = 'blue')
        plt.title("Sales vs City", fontsize = 20, color = 'red')
        plt.xticks(rotation = 45)
        plt.show()
```

```python
widgets.interact(graphs, c2 = c1)
```
